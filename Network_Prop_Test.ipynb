```python
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
```

```python
g = nx.Graph([[2,1],[1,3],[2,4],[1,4]])
nx.draw(g)
```

```python
def propagate_nx(g,p,new_active):
    
    targets = []
    for node in new_active:
        targets += g.neighbors(node)

    return(targets)
```

```python
g_new = propagate_nx(g,0.2,[1,3,4])
g_new
```
